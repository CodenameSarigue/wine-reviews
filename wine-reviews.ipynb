```python
#Removendo as colunas que não serão analisadas. Foi utilizada a função drop(). A coluna Descrição foi removida anteriormente com o comando del
df.drop(columns=["Designação", "Descrição", "Província", "Região_1", "Região_2", "Twitter do degustador", "Vinícola"], inplace=True)
```

```python
#Número de Avaliações de Vinhos por País
import matplotlib.pyplot as plt
df["País"].value_counts().sort_values(ascending=True).plot.barh(figsize=(10,20),title="Avaliações de Vinhos por País", color="grey");
plt.xlabel("Nº de Avaliações");
plt.ylabel("País");
plt.savefig("grafico_aval_vinho_pais.tif")
```

```python
#Número de Variedades de vinho por País
import matplotlib.pyplot as plt
df.groupby("País")["Variedade"].nunique().sort_values(ascending=True).plot.barh(figsize=(10,20), title="Variedades de Vinhos por País", color="grey");
plt.xlabel("Variedade de Vinhos");
plt.ylabel("País");
plt.savefig("grafico_varie_vinho_pais.tif")
```

```python
#Número de Degustadores que avaliaram vinhos de cada País
df.groupby("País")["Nome do degustador"].nunique().sort_values(ascending=True).plot.barh(figsize=(10,20), title="Degustadores por País", color="grey");
plt.xlabel("Nº de Degustadores");
plt.ylabel("Origem do Vinho");
plt.savefig("grafico_degusta_vinho_pais.tif")

```

```python
#Média de pontos por País
round(df.groupby("País")["Pontos"].mean().sort_values(ascending=False), 1)#.plot.barh(figsize=(10,20), title="Pontos por País", color="grey");
#plt.xlabel("Pontos");
#plt.ylabel("País");
#plt.savefig("grafico_pontos_pais.tif")
```

```python
#Verificando os valores nulos.
df.isnull().sum()
```

```python
#Substituindo valores nulos de País para Desconhecido
df["País"].fillna("Desconhecido", inplace=True)
```

```python
#Substituindo valores nulos de Preço pela Média de preço
df["Preço"].fillna(df["Preço"].mean(), inplace=True)
```

```python
#Substituindo valores nulos de Nome do degustador por Desconhecido
df["Nome do degustador"].fillna("Desconhecido", inplace=True)
```

```python
#Substituindo valores nulos de Variedade por Desconhecido
df["Variedade"].fillna("Desconhecido", inplace=True)
```

```python
#Verificando os valores nulos. Tudo ok após substituir os valores nulos por Desconhecido e Preços desconhecidos pela Média de preço nos quatro códigos supracitados.
df.isnull().sum()
```

```python
#Criando a coluna Pontos/Preço, informação que poderá nos ajudar a inferir qual os vinhos de melhor custo benefício
df["Pontos/Preço"] = round(df["Pontos"] / df["Preço"], 2)
```
